```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the ticker symbols for the stocks and ETFs
symbols = ["FB", "AAPL", "AMZN", "NFLX", "GOOGL", "TSLA", "SHOP", "NVDA", "ROKU", "ARKK", "ARKG"]

# Specify the date range for the historical data
start_date = "2020-01-01"
end_date = "2023-01-01"

# Create an empty DataFrame to store the closing prices
closing_prices = pd.DataFrame()

# Fetch historical closing prices for each stock and ETF
for symbol in symbols:
    data = yf.download(symbol, start=start_date, end=end_date)
    closing_prices[symbol] = data['Adj Close']

# Create a DataFrame to store S&P 500 data
sp500_data = yf.download('^GSPC', start=start_date, end=end_date)
closing_prices['S&P 500'] = sp500_data['Adj Close']

# Calculate daily percentage returns for all assets
returns = closing_prices.pct_change()

# Remove rows with any missing values (NaN)
cleaned_returns = returns.dropna()

# Set the figure size for the plot
plt.figure(figsize=(14, 8))

# Plot the relative returns for each stock and ETF
for symbol in cleaned_returns.columns:
    if symbol != 'S&P 500':
        plt.plot(cleaned_returns.index, cleaned_returns[symbol] - cleaned_returns['S&P 500'], label=symbol)

# Customize the plot
plt.title('Relative Returns of Selected Ticker Symbols vs. S&P 500', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Relative Returns', fontsize=14)
plt.legend(loc='upper left', fontsize=12, frameon=True)
plt.grid(True, linestyle='--', alpha=0.7)

# Format date ticks
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(prune='both'))

# Show the plot with a consistent y-axis scale
plt.ylim(-0.3, 0.3)

# Add a horizontal line at y=0 for reference
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

# Show the plot with a tight layout
plt.tight_layout()

# Save the plot as an image (optional)
plt.savefig('relative_returns.png', dpi=300)

# Show the plot
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the ticker symbols for the stocks and ETFs
symbols = ["FB", "AAPL", "AMZN", "NFLX", "GOOGL", "TSLA", "SHOP", "NVDA", "ROKU", "ARKK", "ARKG"]

# Specify the date range for the historical data
start_date = "2010-01-01"
end_date = "2023-01-01"

# Create an empty DataFrame to store the closing prices
closing_prices = pd.DataFrame()

# Fetch historical closing prices for each stock and ETF
for symbol in symbols:
    data = yf.download(symbol, start=start_date, end=end_date)
    closing_prices[symbol] = data['Adj Close']

# Create a DataFrame to store S&P 500 data
sp500_data = yf.download('^GSPC', start=start_date, end=end_date)
closing_prices['S&P 500'] = sp500_data['Adj Close']

# Calculate daily percentage returns for all assets
returns = closing_prices.pct_change()

# Remove rows with any missing values (NaN)
cleaned_returns = returns.dropna()

# Calculate the value of $1 invested in GOOGL since 2010
investment_value = (1 + cleaned_returns['GOOGL']).cumprod()

# Set the figure size for the plot
plt.figure(figsize=(12, 6))

# Plot the investment value
plt.plot(cleaned_returns.index, investment_value, label='Investment Value', color='blue')

# Customize the plot
plt.title('Investment in GOOGL Since 2010', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Investment Value', fontsize=14)
plt.legend(loc='upper left', fontsize=12, frameon=True)
plt.grid(True, linestyle='--', alpha=0.7)

# Format date ticks
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(prune='both'))

# Show the plot with a tight layout
plt.tight_layout()

# Show the plot
plt.show()


```
